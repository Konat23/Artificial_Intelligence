# Regression Lineal in Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Some synthetic data

```python
#Generate some synthetic data  Y= a0+a1*X = b_real + w_real*X
np.random.seed(16)  # the same values

N = 500
X = np.random.normal(50, 100, N)  # np.random.normal(mean, dev standar, size of the data)
noise = 2*np.random.normal(0, 100, N)
Y = np.dot(X,w_real) + b_real + noise
plt.scatter(X,Y)
```

## Functions:

```python
def cost(Y,Y_predict):
  '''
  Parameters:
  ------------------
  Y : vector with the data
  Y_predict : vector with the predictions

  Returns
  ------------------
  cost = The MSE between the original value y and the prediction y_prediction

  '''
  N = Y.shape[0]
  mse1 = np.sum((Y-Y_predict)**2)/N
  return mse1
```

```python
def initialize_model():
  '''
  Initialize the value of w and b

  Returns
  -----------
  w,b: The initial value of w and b
  '''
  np.random.seed(16)  # the same values
  w = np.random.normal(0,2)
  b = np.random.normal(0,2)

  return [w,b]
```

```python
def predict(w,b,X):
  '''
  Parameters:
  ------------------
  w, b : parameters of the lineal model
  x : data
  Returns
  ------------------
  The prediction Y_prediction = w*X + b

  '''
  Y_prediction = w*X + b
  return Y_prediction
```

```python
def gradients(w_current,b_current,alpha,X,Y):
  '''
  Calculate the gradients

  Parameters:
  --------------
  w_current, b_current: the current values of w and b
  alpha: learning rate
  X,Y : Vectors of the Data

  Returns
  ---------------------
  w, b : the update value of w and b
  '''
  N = X.shape[0]

  # Calculating the gradients
  dw = (-2/N)*np.sum((Y-w_current*X-b_current)*X)
  db = (-2/N)*np.sum(Y-w_current*X-b_current)

  # Updating w and b
  w = w_current - alpha*dw
  b = b_current - alpha*db

  return w, b
```

## The initial model

```python
[w,b] = initialize_model() # To initialize the model, i.e. the values of w1 and b.
Y2 = np.dot(X,w) + b
error = cost(Y,Y2)
plt.scatter(X,Y) # Draw the original data
plt.plot(X,Y2,'r') # Draw the prediction
plt.title(f"Error = %.2f" % error)
```

## Fit the model

```python
# Fit the model
# Develop an algorith that using Gradient Descent find the values of w and b that
# minimizes the cost function
Cost = []  # empty list to save the cost in each iterations
alpha = 0.00007

for i in range(30):
   w, b = gradients(w,b,alpha,X,Y)
   Cost.append(cost(Y,predict(w,b,X)))




```

## Final Model

```python
Y2 = np.dot(X,w) + b
error = cost(Y,Y2)
plt.scatter(X,Y) # Draw the original data
plt.plot(X,Y2,'r') # Draw the prediction
plt.title(f"Error = %.2f" % error)
```
