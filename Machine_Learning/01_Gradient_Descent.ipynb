# Main

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0) # Para reproductivilidad
```

```python
# Creamos la funcion propuesta como expresion simbolica y funcion lambda
x = sp.symbols('x')
f = 2*x**4-2*x**3-60*x**2-24*x+10
f_np = sp.lambdify(x,f,'numpy')

# Calculamos la derivada como expresion simbolica y funccion lambda
df = sp.diff(f, x)
df_np = sp.lambdify(x,df,'numpy')
```

```python
# Inicializamos el punto de inicio
x0 = np.random.uniform(-5,5)
y0 = f_np(x0)
```

### Grafica de la funcion y punto inicial

```python
# Graficamos entre -6 y 6 la funcion y mostramos el punto inicial
xv = np.linspace(-6, 6, 100)
yv = f_np(xv)
plt.plot(xv,yv)
plt.plot(x0,y0,marker='D')
```

### Iterar

```python

xi = x0
alpha = 0.001 # definimos el coeficiente de paso del gradiente
iter = 0 # contador de iternaciones
while np.abs(df_np(xi))>0.001: # ejecutar el while hasta tener pendiente peque√±a
  # Mostramos el punto actual
  yi = f_np(xi)
  print(f"Iteracion Numero: {iter}: xi: {xi}, yi {yi}, m: {df_np(xi)}")

  #Calculamos el siguiente punto
  xi = xi - alpha*df_np(xi)
  iter+=1

print(f"Iteracion Final: {iter}: xi: {xi}, yi {f_np(xi)}, m: {df_np(xi)}")
```

### Grafica valor final

```python
# Graficamos entre -6 y 6 la funcion y mostramos el punto final.
xv = np.linspace(-6, 6, 100)
yv = f_np(xv)
plt.plot(xv,yv) # Evaluar la funcion en el rango
plt.plot(xi,yi,marker='D') # grafica del punto final
```
